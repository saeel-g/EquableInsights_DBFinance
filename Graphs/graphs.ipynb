```python

p1=df1[df1['Plan_Shorthand']=='Arizona EORP']
p1=p1[["Plan_Shorthand",'Year','BenefitsPerPerson']]

p2=df1[df1['Plan_Shorthand']=='Arizona PSPRS']
p2=p2[["Plan_Shorthand",'Year','BenefitsPerPerson']]

p3=df1[df1['Plan_Shorthand']=='Arizona SRS']
p3=p3[["Plan_Shorthand",'Year','BenefitsPerPerson']]

p4=df1[df1['Plan_Shorthand']=='Phoenix ERS']
p4=p4[["Plan_Shorthand",'Year','BenefitsPerPerson']]

p5=df1[df1['Plan_Shorthand']=='Arizona CORP']
p5=p5[["Plan_Shorthand",'Year','BenefitsPerPerson']]

p6=df1[df1['Plan_Shorthand']=='Arizona PSPRS Tier 3']
p6=p6[["Plan_Shorthand",'Year','BenefitsPerPerson']]

p7=df1[df1['Plan_Shorthand']=='Tucson Supplemental RS']
p7=p7[["Plan_Shorthand",'Year','BenefitsPerPerson']]
```

```python

data = pd.concat([p1,p2,p3,p4,p5,p6,p7], ignore_index=True)
```

```python
sns.set_style("whitegrid")

# Create the benefits vs. year plot with different plans
plt.figure(figsize=(12,8))
sns.lineplot(x='Year', y='BenefitsPerPerson', data=data, hue='Plan_Shorthand')
for plan in data['Plan_Shorthand'].unique():
    plan_data = data[data['Plan_Shorthand'] == plan]
    for _, row in plan_data.iterrows():
        plt.scatter(row['Year'], row['BenefitsPerPerson'], color='skyblue', marker='o')
# Customize the plot
plt.xlabel('Year')
plt.ylabel('Benefits')
plt.title('BenefitsPerPerson vs. Years for Different Plans for Pension Plan Schemes in Arizona')
plt.legend(title='Plan')

# Show the plot
plt.show()
```

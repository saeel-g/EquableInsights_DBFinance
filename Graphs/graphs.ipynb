```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
```

```python
df=pd.read_csv('for_graph.csv')
```

```python
df['StateName'].value_counts()
```

```python
df1=df[df['StateName']=='Arizona']
df1.drop('Unnamed: 0',axis=1)
df1['Plan_Shorthand'].value_counts()
```

```python
df1['Plan_Shorthand'].value_counts()
```

```python
df2=df[df['StateName']=='Missouri']
df2.drop('Unnamed: 0',axis=1)
df2['Plan_Shorthand'].value_counts()
```

```python

p1=df1[df1['Plan_Shorthand']=='Arizona EORP']
p1=p1[["Plan_Shorthand",'Year','BenefitsPerPerson']]

p2=df1[df1['Plan_Shorthand']=='Arizona PSPRS']
p2=p2[["Plan_Shorthand",'Year','BenefitsPerPerson']]

p3=df1[df1['Plan_Shorthand']=='Arizona SRS']
p3=p3[["Plan_Shorthand",'Year','BenefitsPerPerson']]

p4=df1[df1['Plan_Shorthand']=='Phoenix ERS']
p4=p4[["Plan_Shorthand",'Year','BenefitsPerPerson']]

p5=df1[df1['Plan_Shorthand']=='Arizona CORP']
p5=p5[["Plan_Shorthand",'Year','BenefitsPerPerson']]

p6=df1[df1['Plan_Shorthand']=='Arizona PSPRS Tier 3']
p6=p6[["Plan_Shorthand",'Year','BenefitsPerPerson']]

p7=df1[df1['Plan_Shorthand']=='Tucson Supplemental RS']
p7=p7[["Plan_Shorthand",'Year','BenefitsPerPerson']]
```

```python
p1_2=df2[df2['Plan_Shorthand']=='Missouri DOT']
p1_2=p1_2[["Plan_Shorthand",'Year','BenefitsPerPerson']]

p2_2=df2[df2['Plan_Shorthand']=='Missouri LGERS']
p2_2=p2_2[["Plan_Shorthand",'Year','BenefitsPerPerson']]
                                                         
p3_2=df2[df2['Plan_Shorthand']=='Missouri PEERS']
p3_2=p3_2[["Plan_Shorthand",'Year','BenefitsPerPerson']]

p4_2=df2[df2['Plan_Shorthand']=='Missouri PSRS']
p4_2=p4_2[["Plan_Shorthand",'Year','BenefitsPerPerson']]

p5_2=df2[df2['Plan_Shorthand']=='Missouri SERS']
p5_2=p5_2[["Plan_Shorthand",'Year','BenefitsPerPerson']]

p6_2=df2[df2['Plan_Shorthand']=='Kansas City Missouri ERS']
p6_2=p6_2[["Plan_Shorthand",'Year','BenefitsPerPerson']]

p7_2=df2[df2['Plan_Shorthand']=='Kansas City Missouri Schools']
p7_2=p7_2[["Plan_Shorthand",'Year','BenefitsPerPerson']]

p8_2=df2[df2['Plan_Shorthand']=='St. Louis School Employees']
p8_2=p8_2[["Plan_Shorthand",'Year','BenefitsPerPerson']]
```

```python

data1 = pd.concat([p1,p2,p3,p4,p5,p6,p7], ignore_index=True)
data2 = pd.concat([p1_2,p2_2,p4_2,p5_2,p7_2,p8_2], ignore_index=True)
```

```python
plt.figure(figsize=(12, 8))
sns.set_style("whitegrid")

# Assuming 'data1' contains your data
sns.lineplot(x='Year', y='BenefitsPerPerson', data=data2, hue='Plan_Shorthand')
for plan in data2['Plan_Shorthand'].unique():
    plan_data = data2[data2['Plan_Shorthand'] == plan]
    for _, row in plan_data.iterrows():
        plt.scatter(row['Year'], row['BenefitsPerPerson'], color='skyblue', marker='o')
# Add a dollar sign to the y-axis labels
def format_dollar(x, pos):
    return f"${x:.2f}"

# Apply the formatter to the y-axis
plt.gca().yaxis.set_major_formatter(mticker.FuncFormatter(format_dollar))

# Customize the plot
plt.xlabel('Year')
plt.ylabel('Benefits Per Person')
plt.title('Benefits Per Person vs. Years for Different Plans for Pension Plan Schemes in Missouri')
plt.legend(title='Plan')

# Save the plot to an image file
plt.savefig('Missouri_State.png', bbox_inches='tight')  # Adjust the filename and format as needed
# Show the plot (optional)
plt.show()
```
